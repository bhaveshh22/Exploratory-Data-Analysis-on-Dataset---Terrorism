
# Plotting a Histogram

```python
data.hist(figsize=(100,110))
plt.show() 
```

```python
sns.set(rc={'figure.figsize':(25,40)})
sns.countplot(x=data['iyear'],hue=data['region_txt'])
plt.title('Region_wise terrorist activity in each year')
```

```python
pd.crosstab(data['iyear'],data['region_txt']).plot(kind='line',figsize=(15,20))
plt.title('Region-wise terrorist activity in each year')
plt.ylabel('Number od Attacks')
plt.xlabel('years')
plt.show()

print('Middle east & North Africa has been reported to have more incidents')
```

```python
data['region_txt'].value_counts()
```

```python
data['country_txt'].value_counts()
```

```python
plt.figure(figsize=(15,6))
sns.barplot(data['country_txt'].value_counts()[:10].index,data['country_txt'].value_counts()[:10].values)
plt.title('Top 10 countries effected by Attacks',fontsize=15)
plt.ylabel('counts',fontsize=15)
plt.xlabel('countries',fontsize=15)
plt.show()

print('Most effected country is Iraq & mostly belongs to Asia countinent')
```

```python
plt.figure(figsize=(15,6))
sns.barplot(data['city'].value_counts()[:10].index,data['city'].value_counts()[:10].values)
plt.title('Top 10 city effected by Attacks',fontsize=15)
plt.ylabel('counts',fontsize=15)
plt.xlabel('city',fontsize=15)
plt.show()

print('Most effected in baghdad & unknown')
```

```python
sns.set(rc={'figure.figsize':(30,7)})
sns.countplot(x=data['region_txt'],hue=data['weaptype1_txt'])
plt.title('Weapon type used in per region')
plt.xlabel('Region',fontsize=15)
print('Explosives,Chemicals nad Firearms and widely used as a weapon type')
```

```python
sns.set(rc={'figure.figsize':(30,7)})
sns.countplot(x=data['region_txt'],hue=data['attacktype1_txt'])
plt.title('Attack type used in per region')
plt.xlabel('Region',fontsize=15)
print('Boobing/Explosion is highly used attack type nd region is Middle East/North Africa')
```

```python
sns.set(rc={'figure.figsize':(30,7)})
sns.countplot(x=data['region_txt'],hue=data['targtype1_txt'])
plt.title('Attack type used in per region')
plt.xlabel('Region',fontsize=15)
print('Private Citizens/Property is the highley target type in each region')
```

```python
data.plot(kind='scatter',x='nkill',y='nwound',color='green',figsize=(10,10),fontsize=15)
plt.xlabel('Kill')
plt.ylabel('Wound')
plt.title('Kill_wound Scatter plot')
plt.show()
```
